## 1. Dataset
### 1.1 Upload Dataset

The first step is to upload the dataset. It comes from the Kaggle database and is called the 'Predicting Pulsar Star Dataset'. Each variable pertains to an integrated pulsar profile, representing, for example, the standard deviation. The descriptions of the variables will be provided below.

An integrated pulsar profile (also known as a folded profile) is a representation of a neutron star's signal, observed as regular pulsations.The pulsar profile is generated by monitoring and recording the radiation intensity from the neutron star over a specific time span. However, due to the star's rotational motion, pulsations may occur at different rotational phases. To consolidate these pulsations and obtain a more representative image, a folded pulsar profile is created by averaging the signal in both time and frequency dimensions.

The folded pulsar profile contains information about the shape and features of the neutron star's pulsations, such as width, asymmetry, amplitude, or the number of pulses. The statistics obtained from this profile, mentioned earlier, provide additional numerical data that describe these features in a more condensed and analyzable manner.

### 1.2 Data analysis

Dataset contains 9 variables. 

 - target_class - is the signal Pulsar or not, binary
 - Mean of the integrated profile - average intensity of the pulsar signal across the observation period
 - Standard deviation of the integrated profile - quantity of how much the intensity values deviate from the mean value
 - Excess kurtosis of the integrated profile - information about the shape of the distribution of intensity values
 - Skewness of the integrated profile - the degree to which the intensity values are skewed or shifted to one side of the mean

DM-SNR curve -  plot that shows the relationship between the Dispersion Measure (DM) and the Signal-to-Noise Ratio (SNR) for a collection of pulsar candidates. In a DM-SNR curve, each data point represents a pulsar candidate or a detected pulsar, with the DM value on the x-axis and the SNR value on the y-axis.
DM -  total electron content along the line of sight from the pulsar to the observer
SNR -  measure of the strength of a signal relative to the background noise

 - Mean of the DM-SNR curve - average strength of the pulsar candidates or detected pulsars in the dataset
 - Standard deviation of the DM-SNR curve - indicates how much the SNR values deviate from the mean SNR value
 - Excess kurtosis of the DM-SNR curve - distribution and characteristics of the SNR values of pulsar candidates
 - Skewness of the DM-SNR curve - asymmetry in the distribution of SNR values for a set of pulsar candidates or detected pulsars, plotted against their corresponding DM values on the DM-SNR curve

#### NaN values treatment

```python
imputer = SimpleImputer(strategy='mean')

train = df_train.copy()
train[df_train.columns] = imputer.fit_transform(df_train[df_train.columns])

train.isnull().sum()
```

Since not all models accept NaN values, we have imputed them using the median strategy. For instance, the feature "Excess kurtosis of the integrated profile" had numerous NaN values and a high correlation score. However, we chose not to drop these columns, as correlation is also considered when comparing them with the "target_class."

#### Transposition of features

```python
print(train.columns)
```

```python
train_tr = train.groupby('target_class')[['Mean of the integrated profile',
       'Standard deviation of the integrated profile',
       'Excess kurtosis of the integrated profile',
       'Skewness of the integrated profile', 'Mean of the DM-SNR curve',
       'Standard deviation of the DM-SNR curve',
       'Excess kurtosis of the DM-SNR curve', 'Skewness of the DM-SNR curve']].mean().reset_index()

train_tr= train_tr.transpose().reset_index()
train_tr.columns = ['features', 'Not Pulsar Star', 'Pulsar Star']
```

```python
X = train.drop('target_class',axis=1)
y = train['target_class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1)

ss=StandardScaler() 
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

Splitting the dataset and scaling it using a standard scaler to mitigate the influence of the class 0.

### 2.2 Building models

#### Linear Regression

```python
lr_model = LogisticRegression(random_state=42,solver="saga",C=0.5,penalty="l2")

lr_model.fit(X_train,y_train)

lr_pred = lr_model.predict(X_test)

lr_score = lr_model.score(X_test,y_test)
print(lr_score)
print(classification_report(y_test, lr_pred))
```

Search Grid - Linear Regression

```python
param_grid = {
    'fit_intercept': [True, False],
    'solver': ['saga', 'newton-cholesky', 'lbfgs', 'sag', 'liblinear', 'newton-cg']
}

grid_search = GridSearchCV(lr_model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("Best parameters: ")
print(best_model.get_params())
lr_score = lr_model.score(X_test,y_test)
print(f'Grid Search lr_model score: {lr_score}')

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(classification_report(y_test, y_pred))
report_lr = classification_report(y_test, y_pred, output_dict=True)
```

#### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train,y_train)

dt_pred = dt_model.predict(X_test)


dt_score = dt_model.score(X_test,y_test)
print(dt_score)
print(classification_report(y_test, dt_pred))
```

Search Grid - Decision Tree Classifier

```python
param_grid = {
    'criterion': ['gini', 'entropy'],  
    'max_depth': [None, 10, 20, 30],    
    'min_samples_split': [2, 5, 10],    
    'min_samples_leaf': [1, 2, 4]       
}
grid_search = GridSearchCV(dt_model,param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("Best parameters: ")
print(best_model.get_params())
dt_score = dt_model.score(X_test,y_test)
print(f'Grid Search dt_model score: {dt_score}')

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(classification_report(y_test, y_pred))
report_dt = classification_report(y_test, y_pred, output_dict=True)
```

#### SVM

```python
svm_model = SVC(kernel='linear', random_state=42)

svm_model.fit(X_train, y_train)

svm_pred = svm_model.predict(X_test)

print(classification_report(y_test, svm_pred))
```

Search Grid - SVM

```python
param_grid = {
    'C': [0.1, 1, 10],             
    'kernel': ['linear', 'rbf'],   
    'gamma': ['scale', 'auto'],    
    'class_weight': [None, 'balanced']  
}

grid_search = GridSearchCV(svm_model,param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("Best parameters: ")
print(best_model.get_params())
svm_score = svm_model.score(X_test,y_test)
print(f'Grid Search svm_model score: {svm_score}')

mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(classification_report(y_test, y_pred))
report_svm = classification_report(y_test, y_pred, output_dict=True)
```
